```python
# Import necessary libraries
import keras_cv
import matplotlib.pyplot as plt
```

```python
# Construct a model
model = keras_cv.models.StableDiffusion(
    img_width=512, img_height=512, jit_compile=False
)
```

```python
# Function to plot generated images
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")
```

```python
prompt = "surreal dream of a magical underwater world, glowing sea creatures, vibrant colors, ethereal atmosphere, mystery and wonder"
images = model.text_to_image(prompt, batch_size=3)

# Plot the generated images
plot_images(images)
plt.show()
```
